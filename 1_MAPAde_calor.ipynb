```python
# Cargar el archivo (ajusta el nombre si es necesario)
import pandas as pd
df = pd.read_csv("Datos1 - Final.csv", encoding='latin1')  # <-- reemplaza con el nombre real
df.head()
```

```python
correlation_matrix = df[['precipitacion', 'produccion_agricola', 'energiapercapita']].corr()
print("\n游늵 Matriz de correlaci칩n:")
print(correlation_matrix)
```

```python
# 游늳 Mostrar la matriz como un mapa de calor
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaci칩n - Variables WEFE')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x='energiapercapita', y='produccion_agricola', data=df)
plt.title('Relaci칩n entre Consumo Energ칠tico y Producci칩n Agr칤cola')
plt.xlabel('Consumo Energ칠tico (kWh)')
plt.ylabel('Producci칩n Agr칤cola (toneladas)')
plt.show()
```
